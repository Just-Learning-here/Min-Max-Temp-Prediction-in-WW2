```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv(r"C:\Users\bdr-m\Downloads\Weather Condition\Summary of Weather.csv")

''' this project should be able to answer these Questions : 
    Is there a relationship between the daily minimum and maximum temperature? 
    Can you predict the maximum temperature given the minimum temperature?
'''
print('')
```

```python
#display(df)
#df.info()
# what i need from this data is MaxTemp and MinTemp 
predictWeather=df[['MaxTemp','MinTemp']]
minTemp=predictWeather[['MinTemp']]
maxTemp=predictWeather[['MaxTemp']]
display(predictWeather)
```

```python
plt.scatter(minTemp,maxTemp, color='red', marker='+')
plt.xlabel('Minimum Temperature in Celsius')
plt.ylabel('Maximum Temperature in Celsius')
```

```python

#predictWeather.MaxTemp
reg= linear_model.LinearRegression()
minTemp= minTemp.values
reg.fit(minTemp,maxTemp)
print("Answers to the questions asked prior to the analysis: ")
print("Q1:Yes we can as we saw from the plot that there is a linear realtionship between The min temp and The max ")
print("Q2: Yes based on the linear Equation ")
```
